#### Load the tokenizer and dataset

Here we're loading the AutoTokenizer class from the 🤗 Transformers library.

In this notebook, I'm applying preprocessing (tokenzing) on the GLUE SST-2 dataset. It’s a little bit different since it’s composed of single sentences instead of pairs.


```python
from transformers import AutoTokenizer
from datasets import load_dataset

raw_datasets = load_dataset("glue", "sst2")
raw_datasets
```

```python
raw_train_dataset = raw_datasets["train"]
raw_train_dataset[0]
```

```python
raw_train_dataset.features
```

Define the Tokenize function

```python
checkpoint = "bert-base-cased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
tokenizer
```

```python
def tokenize_function(example):
  return tokenizer(example["sentence"], truncation=True)
```

### Tokenize the dataset using the map function

```python
tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)
tokenized_datasets
```
